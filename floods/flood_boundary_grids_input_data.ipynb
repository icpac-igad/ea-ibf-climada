## Grid creation

```python
#!/usr/bin/python

import argparse
import pandas as pd
from shapely.geometry import Polygon, mapping
import simplekml
import numpy as np
import itertools
import geopandas as gp
import ntpath
import math

import shapely.wkt

# Instantiate the parser
"""
python multiple_gridkml_creator_v1.py gridextent_coords.csv

gridextent_coords.csv is input file having cordinate extents


"""

def map_extent(plot_extent): 
    geo_fe= Polygon([[plot_extent['w'],plot_extent['s']], [plot_extent['w'],plot_extent['n']], [plot_extent['e'],plot_extent['n']],[plot_extent['e'],plot_extent['s']]])
    return geo_fe

def crs_find_centroid(swlong,swlat,gridx,gridy,nx,ny):
    nelong=swlong+(gridx*nx)
    nelat=swlat+(gridy*ny)
    plot_extent={}
    plot_extent['w'],plot_extent['s'],plot_extent['n'],plot_extent['e']=swlong,swlat,nelat,nelong
    geo_extent=map_extent(plot_extent)
    cen_map_extent=geo_extent.centroid
    lat=cen_map_extent.y
    lon=cen_map_extent.x
    utm_band = str((math.floor((lon + 180) / 6 ) % 60) + 1)
    if len(utm_band) == 1:
        utm_band = '0'+utm_band
    if lat >= 0:
        epsg_code = '326' + utm_band
    else:
        epsg_code = '327' + utm_band
    return epsg_code


def gridcreator(swlong,swlat,gridx,gridy,nx,ny):
    """
    creates polygon grids as a geometry list for the speciifed variables

    Parameters
    ----------
    swlong : south west longitude value in degree decimals (float)
    swlat : south west latitude value in degree decimals (float)
    gridx : width size of the grid required in degress  (float)
    gridy : width size of the grid required in degress  (float)
    nx : number of grids required in x direction (integer)
    ny : number of grids required in y direction (integer)
            
    Returns
    -------
    a list of polygons geometry for the given variables
  
    Notes
    -------
    module for grid computation, effective to southern parts, 
    it really shrinking the grid size while moving to north
    """
    nelong=swlong+(gridx*nx)
    nelat=swlat+(gridy*ny)
    xmin,ymin,xmax,ymax= swlong, swlat, nelong, nelat
    gridWidth = gridx
    gridHeight = gridy
    rows = round((ymax-ymin)/gridHeight)
    cols = round((xmax-xmin)/gridWidth)
    ringXleftOrigin = xmin
    ringXrightOrigin = xmin + gridWidth
    ringYtopOrigin = ymax
    ringYbottomOrigin = ymax-gridHeight
    polygun1=[]
    for i in np.arange(cols):
        ringYtop = ringYtopOrigin
        ringYbottom =ringYbottomOrigin
        for j in np.arange(rows):
            polygon = Polygon([(ringXleftOrigin, ringYtop), (ringXrightOrigin, ringYtop), (ringXrightOrigin, ringYbottom), (ringXleftOrigin, ringYbottom)])
            polygun1.append(polygon)
            ringYtop = ringYtop - gridHeight
            ringYbottom = ringYbottom - gridHeight
        ringXleftOrigin = ringXleftOrigin + gridWidth
        ringXrightOrigin = ringXrightOrigin + gridWidth
    return polygun1   




def maille_counter(grid_df):
    """
    creates a pandas datframe with Maille and Maille_X and Maille_Y columns
    Parameters
    ----------
    grid_df : a pandas dataframe with polygon geometries
            
    Returns
    -------
    a pandas dataframe with Maille columns
    """
    grid_df.columns=['geometry']
    grid_df['gn']=np.arange(1,len(grid_df.index)+1,1)
    wef=grid_df['gn'].to_numpy()
    # it's a gymnastics to get sw count for grids
    o=wef.reshape(nx, ny)
    p=o.T
    q=np.ravel(p,order='F')
    b=wef.reshape(ny, nx)
    c=b.T
    d=np.rot90(c, 1)
    e=np.ravel(d,order='F')
    qq=pd.DataFrame()
    qq['gn']=q
    qq['Maille']=e
    qq1=qq.sort_values('Maille')
    ww1=pd.merge(grid_df,qq1)
    #this is for giving the column(Maille_X) and row(Maille_Y) numbering
    rows=np.arange(1,ny+1,1)
    cols=np.arange(1,nx+1,1)
    rows1=[rows] * nx
    rows2=[item for sublist in rows1 for item in sublist]
    rows3=np.fliplr([rows2])[0]
    cols1=list(itertools.chain.from_iterable(itertools.repeat(x,ny) for x in cols))
    ww1['Maille_Y']=rows3
    ww1['Maille_X']=cols1
    return ww1



def xy_maker(maille_grid,epsg_code):
    """
    creates a geopandas dataframe with centroid, lower end and upper end cordinate details
    calculates grid area

    Parameters
    ----------
    maille_grid : Dataframe with Maille grids
            
    Returns
    -------
    Geopandas dataframe
    """
    maille_grid['centroid']=maille_grid['geometry'].map(lambda x: x.centroid)
    maille_grid["X"] = maille_grid.centroid.map(lambda p: p.x)
    maille_grid["Y"] = maille_grid.centroid.map(lambda p: p.y)
    maille_grid['X1']=maille_grid['geometry'].map(lambda x: x.exterior.coords[3][0])
    maille_grid['Y1']=maille_grid['geometry'].map(lambda x: x.exterior.coords[3][1])
    maille_grid['X2']=maille_grid['geometry'].map(lambda x: x.exterior.coords[1][0])
    maille_grid['Y2']=maille_grid['geometry'].map(lambda x: x.exterior.coords[1][1])
    maille_grid.drop(['centroid','gn'], axis=1, inplace=True)
    maille_grid1=maille_grid.sort_values('Maille')
    maille_grid2=gp.GeoDataFrame(maille_grid1)
    maille_grid2.crs = {'init' :'epsg:4326'}
    maille_grid2['area_cell']=maille_grid2['geometry'].to_crs({'init': 'epsg:{0}'.format(epsg_code)})\
               .map(lambda p: p.area / 10**6)
    #maille_grid2['area_cell']=maille_grid2['geometry'].map(lambda p: p.area  / 1e+6 )
    maille_grid3=maille_grid2.reset_index()
    maille_grid3.drop(['index'], axis=1, inplace=True)
    return maille_grid3

def path_leaf(path):
    """
    Get the name of a file without any extension from given path

    Parameters
    ----------
    path : file full path with extension
    
    Returns
    -------
    str
       filename in the path without extension

    """
    head, tail = ntpath.split(path)
    return tail or ntpath.basename(head)

def shp_maker(ue_grid,filepath):
    """
    creates a kml file with onclick baloon details on Maille and cordinates

    Parameters
    ----------
    ue_grid: dataframe
    filepath: kml file name and path
            
    Returns
    -------
    null
    """
    shpfilename=filepath
    csvfilename=filepath.replace('.shp','.csv')
    float_formatcolumns=['X', 'Y', 'X1', 'Y1','X2', 'Y2', 'area_cell']
    #ue_grid=ue_grid0.round(3)
    #print(ue_grid.head())
    ue_grid_p1=ue_grid.drop(float_formatcolumns, axis=1)
    ue_grid_p2=ue_grid[float_formatcolumns].applymap(lambda x: '{0:.3f}'.format(x))
    ue_grid=ue_grid_p1.join(ue_grid_p2)
    ue_grid['geometry']= ue_grid['geometry'].apply(lambda x: shapely.wkt.loads(shapely.wkt.dumps(x, rounding_precision=3)))
    ue_grid.to_file(shpfilename,index=False)
    #ue_grid.to_file(shpfilename,index=False)
    ue_grid1=ue_grid.drop(['geometry'], axis=1)
    ue_grid2=ue_grid1[['Maille', 'Maille_X', 'Maille_Y', 'X1', 'Y1', 'X2', 'Y2', 'X', 'Y','area_cell']]
    ue_grid2.info()
    ue_grid2.to_csv(csvfilename,index=False, float_format='%.3f') 
```

```python
import rasterio

from rasterio.warp import calculate_default_transform, reproject, Resampling

tiff_input='/home/bulbul/Documents/work_data/flood_itzi/ea_dem/n05e030_dem.tif'

tiff_output='/home/bulbul/Documents/work_data/flood_itzi/ssd_sample_run/ssd_dem.tif'


dst_crs = 'EPSG:3857'
with rasterio.open(tiff_input) as src:
    transform, width, height = calculate_default_transform(
    src.crs, dst_crs, src.width, src.height, *src.bounds)
    kwargs = src.meta.copy()
    kwargs.update({
       'crs': dst_crs,
       'transform': transform,
       'width': width,
       'height': height,
       'nodata':0
       })
    with rasterio.open(tiff_output, 'w+', **kwargs) as dst:
          reproject(
             source=rasterio.band(src,1),
             destination=rasterio.band(dst,1),
             src_transform=src.transform,
             src_crs=src.crs,
             dst_transform=transform,
             dst_crs=dst_crs,
             resampling=Resampling.nearest)
```

## rainfall update for ITZI 

```python
import json
from datetime import timedelta, datetime

import numpy as np
import json

def round_floats(o,digits):
    if isinstance(o, float): return round(o, digits)
    if isinstance(o, dict): return {k: round_floats(v,digits) for k, v in o.items()}
    if isinstance(o, (list, tuple)): return [round_floats(x,digits) for x in o]
    return o


class flood_simulation_params:
    # takes today's date and checks api for today and next 3 days

    daytday = datetime.today() - timedelta(days=1)
    
    

    # now we have all available dates of WRF jsons, in a list,in required datetime format for calling API, for a given range of 4 days including today

    #date = dates
    #dates_list=['20201011']

    # default start_hr is url is 00

    start_hr = '00'

    # assign a threshold in 'mm' of rainfall. this will later checked with raster files

    rainfall_threshold_6hr_mm = 1
    threshold = rainfall_threshold_6hr_mm

    # assign "km" for creating bounding box around a given point, ins for km due East , out for km due West

    ins = 10
    out = 1

    # assign paths for various required files or expected folder creations
    json_path = '/home/bulbul/Documents/work_data/flood_itzi/Flood_modeling/itzifloodsim/input_data/portloc.json'
    input_data = "/home/bulbul/Documents/work_data/flood_itzi/Flood_modeling/itzifloodsim/input_data/"
    bboxpath = "/home/itzifloodsim/rain/"

    # region name for final json output

    region='ea'

    # keep json file of port lat long ready to be called in "portjson" variable
    with open(json_path, encoding='utf-8') as json_data:
         portjson = json.load(json_data, strict=False)
    
    
def port_extent_idx_create(params,date_str):
    with open(f'{params.input_data}{date_str}.json', encoding='utf-8') as json_data:
         wrfjson = json.load(json_data, strict=False)
    #collect wrf extent and grid variables
    x_min=wrfjson['results']['precipitation'][0]['ll_lon']
    y_min=wrfjson['results']['precipitation'][0]['ll_lat']
    nx=wrfjson['results']['precipitation'][0]['nx']
    ny=wrfjson['results']['precipitation'][0]['ny']
    xgrid=wrfjson['results']['precipitation'][0]['resolution_gridx']
    ygrid=wrfjson['results']['precipitation'][0]['resolution_gridy']
    x_max=x_min+(xgrid*nx)
    y_max=y_min+(ygrid*ny)
    x = np.linspace(x_min, x_max, nx)
    y = np.linspace(y_min, y_max, ny)
    #for loop into port location and find grid indeces for each port flood simulation domain extent
    new_port=[]
    for i in params.portjson['features']:
        pos = [i['properties']['latitude'],i['properties']['longitude']]
        # after querying lat long from portjson file, ignore any ports below given lat long , at the moment 15.7, which is the border of MH state
        if (float(pos[0]) < float(15.7)):
            continue
        else:
            # given the point lat long for each port, create a bounding box
            # with given ins out from conf
            ll_lat = pos[0] - (0.008998 * params.out)
            ur_lat = pos[0] + (0.008998 * params.ins)
            ll_lon = pos[1] - (0.008998 * params.out)
            ur_lon = pos[1] + (0.008998 * params.ins)
            i['properties']['x_min']=ll_lon
            i['properties']['y_min']=ll_lat
            i['properties']['x_max']=ur_lon
            i['properties']['y_max']=ur_lat
            i['properties']['resolution_gridx']=xgrid
            i['properties']['resolution_gridy']=ygrid
            i['properties']['start_x']=np.argmin(np.abs(np.array(x)-ll_lon))
            if np.argmin(np.abs(np.array(x)-ll_lon))==np.argmin(np.abs(np.array(x)-ur_lon)):
                i['properties']['end_x']=np.argmin(np.abs(np.array(x)-ur_lon))+2
            else:
                i['properties']['end_x']=np.argmin(np.abs(np.array(x)-ur_lon))+1
            i['properties']['start_y']=np.argmin(np.abs(np.array(y)-ll_lat))
            if np.argmin(np.abs(np.array(y)-ll_lat))==np.argmin(np.abs(np.array(y)-ur_lat)):
                i['properties']['end_y']=np.argmin(np.abs(np.array(y)-ur_lat))+2
            else:
                i['properties']['end_y']=np.argmin(np.abs(np.array(y)-ur_lat))+1
            i['properties']['nx']=nx
            i['properties']['ny']=ny
            new_port.append(i)
    cont_new_port={}        
    cont_new_port['features']=new_port
    return cont_new_port
    
def port_extent_rain_sixhr(params,update_port_extent_json,date_str):
    with open(f'{params.input_data}{date_str}.json', encoding='utf-8') as json_data:
         wrfjson = json.load(json_data, strict=False)
    hr_24=np.r_[ 0:23+1 ]
    hr_six_24=np.array_split(hr_24,4)
    new_port=[]
    for ni in update_port_extent_json['features']:
        ni['properties']['rain']={}
        ni['properties']['rain'][date_str]={}
        for hr_six in hr_six_24:
            hr_six_cont=[]
            for hr_single in hr_six:
                rain_2d=np.array(wrfjson['results']['precipitation'][hr_single]['data_array'])
                port_rain=rain_2d[ni['properties']['start_y']:ni['properties']['end_y']+1,ni['properties']['start_x']:ni['properties']['end_x']+1]
                #port_rain_sum=port_rain.sum(axis=(0, 1))
                port_rain_sum=np.sum(port_rain)
                port_rain_sum_float=float(port_rain_sum)
                hr_six_cont.append(port_rain_sum_float)
            rounded_hr_six_cont=round_floats(sum(hr_six_cont),2)
            ni['properties']['rain'][date_str][hr_six[0]]=rounded_hr_six_cont
        new_port.append(ni)
    cont_new_port={}        
    cont_new_port['features']=new_port
    return cont_new_port

    
    
date_str='20201011'  
params=flood_simulation_params

update_port_extent_json=port_extent_idx_create(params,date_str)

rain_update_port=port_extent_rain_sixhr(params,update_port_extent_json,'20201011')  
```

```python

import grass.script.setup as gsetup


import json
from datetime import timedelta, datetime

import os

from grass_session import Session
import grass.script as gcore
import numpy as np


class flood_simulation_params:
    daytday = datetime.today() - timedelta(days=1)
    start_hr = '00'
    # assign "km" for creating bounding box around a given point, ins for km due East , out for km due West
    ins = 10
    out = 1
    region='ea'
    # assign paths for various required files or expected folder creations
    input_data = "/home/itzifloodsim/input_data/"
    grass_data = "/home/itzifloodsim/grass_data/"




def grass_static_data_upload(params):                                                                                                                   
    region_name_folder=f'{params.grass_data}{params.region}/'
    with Session(gisdb=region_name_folder, location="loc1",create_opts="EPSG:3857"):
        gcore.run_command("r.import", input=f'{params.input_data}ssd_dem.tif', output='ssd10dem', overwrite=True)
        gcore.run_command("r.import", input=f'{params.input_data}ssd_mrd.tif', output='friction_map', overwrite=True)
        gcore.run_command("r.import", input=f'{params.input_data}ssd_wmsk.tif', output='watermask', overwrite=True)
        gcore.run_command("v.import", input=f'{params.input_data}ssd_poly.shp', output='bbox', overwrite=True)
        
        
def grass_env_setup(params):
    gisdb = os.path.join(os.path.expanduser("/home"), f'itzifloodsim/grass_data/{params.region}/')
    # the following path is the default path on MS Windows
    # gisdb = os.path.join(os.path.expanduser("~"), "Documents/grassdata")
    # specify (existing) location and mapset
    location = "loc1"
    mapset   = "PERMANENT"
    gisbase = '/usr/local/grass'
    # Set GISBASE environment variable
    os.environ['GISBASE'] = '/usr/local/grass'
    # the following not needed with trunk
    os.environ['PATH'] += os.pathsep + os.path.join(gisbase, 'extrabin')
    # add path to GRASS addons
    home = os.path.expanduser("/home")
    os.environ['PATH'] += os.pathsep + os.path.join(home, '.grass7', 'addons', 'scripts')
    # define GRASS-Python environment
    gpydir = os.path.join(gisbase, "etc", "python")
    sys.path.append(gpydir)
    os.environ['GISDBASE'] = gisdb
    if sys.platform.startswith('linux'):
       os.environ['LC_ALL'] = "en_GB.UTF-8"
    return gisbase,gisdb,location,mapset


def grass_region_setup(params):
    region_name_folder=f'{params.grass_data_folder}{params.region}/'
    with Session(gisdb=region_name_folder, location="loc1",create_opts="EPSG:3857"):
        gcore.run_command("g.region", vector='bbox@PERMANENT', save='aoi', overwrite=True)
        gcore.run_command("r.clip", input='ssd10dem', output='dem_clip', flags='r', overwrite=True)
        gcore.run_command("r.clip", input='friction_map', output='fric_map', flags='r', overwrite=True)
        gcore.read_command("g.region", raster='dem_clip', flags='p', save='demreg', overwrite=True)
        gcore.run_command("r.clip", input='watermask', output='watermask_clip',flags='r', overwrite=True)
        

def grass_mask_reclass(params):
    region_name_folder=f'{params.grass_data}{params.region}/'
    with Session(gisdb=region_name_folder, location="loc1",create_opts="EPSG:3857"):
        tmp1 = open(rf'{region_name_folder}temp1', 'w+')
        tmp1.write('''-3 = 3''')
        tmp1.close()
        gcore.run_command('r.reclass', input='watermask_clip', output='init_water', rules=f'{region_name_folder}temp1', overwrite=True)
        gcore.run_command('r.watershed', elevation='dem_clip', threshold=2000, stream='streams', drainage='dirs', convergence=5, overwrite=True)    
        gcore.run_command('r.stream.basins', dir='dirs', stream='streams', basin='basin_main', overwrite=True)
        tmp2 = open(rf'{region_name_folder}temp2', 'w+')
        tmp2.write('''* = 1''')
        tmp2.close()
        gcore.run_command('r.reclass', input='streams', output='sel_streams', rules=f'{region_name_folder}temp2', overwrite=True)
        gcore.run_command('r.stream.basins', dir='dirs', stream='sel_streams', basins='basin1', overwrite=True)
        gcore.run_command('r.mask', rast='basin1', overwrite=True)


def grass_vector_layer(params):
    region_name_folder=f'{params.grass_data}{params.region}/'
    with Session(gisdb=region_name_folder, location="loc1",create_opts="EPSG:3857"):  
        gcore.run_command('r.to.vect', input='basin1', output='boundary1', type='area', overwrite=True)
        gcore.run_command('v.type', input='boundary1', output='boundary_1', to_type='boundary', overwrite=True)
        gcore.run_command('v.type', input='boundary_1', output='boundary', from_type='boundary', to_type='line', overwrite=True)
        # if below value=2 then boundary is open, if 4 then bcvalue parameter to be added
        gcore.run_command('v.to.rast', input='boundary', type='line',output='bctype', use='val', value=2, overwrite=True)

        
        
import grass.script.setup as gsetup

from grass_routines import flood_simulation_params
from grass_routines import grass_static_data_upload
from grass_routines import grass_env_setup
from grass_routines import grass_region_setup
from grass_routines import grass_mask_reclass
from grass_routines import grass_vector_layer        
    
params=flood_simulation_params()
grass_static_data_upload(params)
gisbase, gisdb, location, mapset=grass_env_setup(params)
gsetup.init(gisbase, gisdb, location, mapset)
grass_region_setup(params)
grass_mask_reclass(params)
grass_vector_layer()
```
