```python
import glob
import os
import ntpath

ncfiles=glob.glob('/home/data_folder/sm_tamsat/*.nc')


len(ncfiles)
```

```python
def order_file(file_list):
    filepath=os.path.dirname(file_list[0])
    file_list1 = [ntpath.basename(item).split('.')[0] for item in file_list]
    file_list1.sort()
    sorted_file_list=[f'{filepath}/{item}.nc' for item in file_list1]
    return sorted_file_list



od_ncfiles=order_file(ncfiles)
```
