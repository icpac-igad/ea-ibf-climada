```python
from bs4 import BeautifulSoup
from urllib.request import Request, urlopen
import re
import contextlib

def get_links(location_path):
    req = Request(f'http://{url}/{location_path}/', headers={'User-Agent': 'Mozilla/5.0'})
    print(f'http://url/{location_path}/')
    #req = Request(f'http://{url}/IGAD_Combined_Drought_Indicator/CHIRPS/', headers={'User-Agent': 'Mozilla/5.0'})
    with contextlib.closing(urlopen(req)) as x:
        html_page = x.read()
    #html_page = urlopen(req).read()
        soup = BeautifulSoup(html_page, 'html.parser')
        links = []
        for table in soup.findAll('table'):
            for tr in table.findAll('tr'):
                for a in tr.findAll('a'):
                    flname=a['href']
                    if '.nc' in flname:
                        links.append(flname)
                    else:
                        pass
        return links
                

list_location_path=['IGAD_Combined_Drought_Indicator/CHIRPS',
                   'IGAD_Combined_Drought_Indicator/GPCC',
                   'IGAD_Ensemble_Soil_Moisture_Anomaly',
                   'IGAD_Fraction_of_Absorbed_Photosynthetically_Active_Radiation_Anomalies_fAPAR',
                   'IGAD_Fraction_of_Absorbed_Photosynthetically_Active_Radiation_fAPAR',
                   'IGAD_Precipitation/CHIRPS',
                   'IGAD_Precipitation/GPCC',
                   'IGAD_Standardized_Precipitation_Index_SPI1/CHIRPS',
                   'IGAD_Standardized_Precipitation_Index_SPI1/GPCC',
                   'IGAD_Standardized_Precipitation_Index_SPI12/CHIRPS',
                   'IGAD_Standardized_Precipitation_Index_SPI12/GPCC',
                   'IGAD_Standardized_Precipitation_Index_SPI24/CHIRPS',
                   'IGAD_Standardized_Precipitation_Index_SPI24/GPCC',
                   'IGAD_Standardized_Precipitation_Index_SPI3/CHIRPS',
                   'IGAD_Standardized_Precipitation_Index_SPI3/GPCC',
                   'IGAD_Standardized_Precipitation_Index_SPI48/CHIRPS',
                   'IGAD_Standardized_Precipitation_Index_SPI48/GPCC',
                   'IGAD_Standardized_Precipitation_Index_SPI6/CHIRPS',
                   'IGAD_Standardized_Precipitation_Index_SPI6/GPCC',
                   'IGAD_Standardized_Precipitation_Index_SPI9/CHIRPS',
                   'IGAD_Standardized_Precipitation_Index_SPI9/GPCC']



dict_cont=[]
for location_path in list_location_path:
    flname=get_links(location_path)
    print(flname)
    new_dict={}
    new_dict[location_path]=flname
    dict_cont.append(new_dict)
    print(location_path)
```
